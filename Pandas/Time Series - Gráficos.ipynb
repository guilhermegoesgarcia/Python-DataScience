# Alucar - Analisando as vendas

```python
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%matplotlib inline
from matplotlib import pyplot as plt 

print(sns.__version__)
```

```python
alucar = pd.read_csv('data/data_series/alucar.csv')
alucar.head()
```

```python
print(f'Quantidade de linhas e colunas: {alucar.shape}')
```

```python
print(f'Quantidade de dados nulos : {alucar.isna().sum().sum()}')
```

```python
alucar.dtypes
```

```python
alucar.mes = pd.to_datetime(alucar['mes'])
alucar.dtypes
```

```python
sns.lineplot(x='mes',y='vendas',data=alucar)
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes',y='vendas',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Vendas Alucar de 2017 a 2018', loc='left',fontsize=18)
ax.set_xlabel('Tempo',fontsize=14)
ax.set_ylabel('Vendas (R$)',fontsize=14)
ax=ax
```

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes',y='aumento',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Aumento das vendas Alucar de 2017 a 2018', loc='left',fontsize=18)
ax.set_xlabel('Tempo',fontsize=14)
ax.set_ylabel('Aumento (R$)',fontsize=14)
ax=ax
```

```python
def plotar(titulo,labelx,labely,x,y,dataset):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x,y=y,data=dataset)
    ax.figure.set_size_inches(12,6)
    ax.set_title(titulo, loc='left',fontsize=18)
    ax.set_xlabel(labelx,fontsize=14)
    ax.set_ylabel(labely,fontsize=14)
    ax=ax
```

```python
plotar('Aumento das vendas Alucar de 2017 a 2018','Tempo','Aumento','mes','aumento',alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Aceleração das vendas - Alucar 2017 a 2018','Tempo','Aceleração','mes','aceleracao',alucar)
```

```python
plt.figure(figsize=(16,12))

ax = plt.subplot(3,1,1)
ax.set_title('Análise de Vendas da Alucar de 2017 a 2018',fontsize = 18)
sns.lineplot(x='mes',y='vendas',data=alucar)

plt.subplot(3,1,2)
sns.lineplot(x='mes',y='aumento',data=alucar)

plt.subplot(3,1,3)
sns.lineplot(x='mes',y='aceleracao',data=alucar)
ax=ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    
    plt.figure(figsize=(16,12))

    ax = plt.subplot(3,1,1)
    ax.set_title(titulo,fontsize = 18)
    sns.lineplot(x=x,y=y1,data=dataset)

    plt.subplot(3,1,2)
    sns.lineplot(x=x,y=y2,data=dataset)

    plt.subplot(3,1,3)
    sns.lineplot(x=x,y=y3,data=dataset)
    ax=ax
```

```python
plot_comparacao('mes','vendas','aumento','aceleracao',alucar,'Análise de vendas')
```

## Gráfico de auto correlação :

```python
from pandas.plotting import autocorrelation_plot
```

```python
ax=plt.figure(figsize=(12,6))
ax.suptitle('Correlação das vendas',fontsize=18,x=0.26,y=0.95)
ax=autocorrelation_plot(alucar['vendas'])
ax=ax
```

```python
ax=plt.figure(figsize=(12,6))
ax.suptitle('Correlação das aumento',fontsize=18,x=0.26,y=0.95)
ax=autocorrelation_plot(alucar['aumento'][1:])
ax=ax
```

```python
ax=plt.figure(figsize=(12,6))
ax.suptitle('Correlação das aceleracao',fontsize=18,x=0.26,y=0.95)
ax=autocorrelation_plot(alucar['aceleracao'][2:])
ax=ax
```

## Análisando a sazonalidade

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.weekday_name
vendas_por_dia.head(10)
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday':'Terca', 'Wednesday':'Quarta', 'Thursday':'Quinta', 'Friday':'Sexta', 'Saturday':'Sabado',
       'Sunday':'Domingo'}
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
vendas_por_dia.head()
```

```python
vendas_por_dia.head(14)
```

# Agrupando os dias

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas','aumento','aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose([chocolura['vendas']],freq=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.resid
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'observacao':observacao,
    'tendencia':tendencia,
    'sazonalidade':sazonalidade,
    'ruido':ruido
    
})

resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(resultado.index,'observacao','tendencia','sazonalidade',resultado,'Exemplo de StatesModels')
```
